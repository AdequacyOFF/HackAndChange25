```python
import numpy as np

def entropy(p):
    p = np.asarray(p)
    p = p / p.sum()
    return -(p * np.log(p + 1e-12)).sum()

tmp = df.copy()
tmp["pred"] = tmp["probs"].apply(lambda x: int(np.argmax(x)))
tmp["is_sure"] = tmp["probs"].apply(lambda x: float(np.max(x) > 0.8))
tmp["entropy"] = tmp["probs"].apply(entropy)

cluster_stats = (
    tmp.groupby("centroid_id")
    .agg(
        size=("label", "size"),
        majority_label=("label", lambda s: s.value_counts().index[0]),
        majority_frac=("label", lambda s: s.value_counts().iloc[0] / len(s)),
        share_sure=("is_sure", "mean"),
        mean_entropy=("entropy", "mean"),
    )
    .reset_index()
)

full_stats = cluster_stats.merge(purity_df, on="centroid_id", how="left")
full_stats.head()
```

```python
full_stats[["purity", "share_sure", "mean_entropy"]].corr()
```

```python
bad_clusters = (
    full_stats
    .query("purity < 0.7 & size > 50 & share_sure < 0.6")
    .sort_values("purity")
)

bad_clusters.head()
```

```python
for cid in bad_clusters.head(5)["centroid_id"]:
    display(df[df["centroid_id"] == cid].sample(5)[["label", "text"]])
```

```python
cid2maj = dict(
    zip(full_stats["centroid_id"], full_stats["majority_label"])
)

df["cluster_label"] = df["centroid_id"].map(cid2maj)
overall_acc = (df["cluster_label"] == df["label"]).mean()
print("Cluster-majority accuracy:", overall_acc)

print(
    df.groupby("label")
      .apply(lambda g: (g["cluster_label"] == g["label"]).mean())
)
```

```python
for thr in [0.6, 0.7, 0.8, 0.9]:
    good_cids = purity_df.loc[purity_df["purity"] >= thr, "centroid_id"]
    keep_mask = df["centroid_id"].isin(good_cids)
    share = keep_mask.mean()
    print(f"thr={thr}, keep_share={share:.3f}")
```

```python
df["keep"] = df["centroid_id"].isin(good_cids)
print(df.groupby(["label", "keep"]).size().unstack(fill_value=0))
```
